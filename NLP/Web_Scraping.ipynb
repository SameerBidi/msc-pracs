### Web Scraping

```python
import requests
from bs4 import BeautifulSoup

html = requests.get("https://realpython.github.io/fake-jobs/").text

soup = BeautifulSoup(html, "lxml")
```

```python
jobs = []

for job in soup.select("#ResultsContainer > div.column"):
    desig = job.select("h2.title")[0].text
    company = job.select("h3.subtitle")[0].text
    location = job.select("p.location")[0].text
    posted_date = job.select("time")[0].text

    location = location.replace("\n", "").strip()

    jobs.append({
        "designation": desig,
        "company": company,
        "location": location,
        "posted_date": posted_date
    })

jobs
```

```python
import pandas as pd

df = pd.DataFrame(jobs)

df
```
